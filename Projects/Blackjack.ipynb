```python
import random
from IPython.display import clear_output

suits = ("Hearts","Diamonds","Spades","Clubs")
ranks = ('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace')
values = {'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Jack':10,'Queen':10,'King':10,'Ace':11}

playing = True

class Card:
    
    def __init__(self,rank,suit):
        self.rank = rank
        self.suit = suit

    def __str__(self):
        return f"{self.rank} of {self.suit}"
    
class Deck:
    
    def __init__(self):
        self.deck = []
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(rank,suit))
    
    def __str__(self):
        return f"There are {len(self.deck)} cards in the deck the first three cards are:\n{str(self.deck[0])}, {str(self.deck[1])}, and {str(self.deck[2])}"
        
    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        return self.deck.pop(0)
    
class Hand:
    
    def __init__(self):
        self.cards = []
        self.value = 0
        self.aces = 0
        
    def add_card(self,card):
        self.cards.append(card)
        self.value += values[card.rank]
        if card.rank == 'Ace':
            self.aces += 1
        if self.value > 21 and self.aces > 1:
            self.value -= 10
            self.aces -= 1

    def __str__(self):
        empty_string = ''
        for i in self.cards:
            empty_string += str(i) + ' '
        return empty_string
    
class Chips:
    
    def __init__(self):
        self.total = 100
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
        self.bet = 0 
    
    def lose_bet(self):
        self.total -= self.bet
        self.bet = 0 
        
def take_bet(chips):
    '''
    Takes in a chips object, asks for the disired bet, and queries if that is possible.
    '''
    while True:
        print(f"You have {chips.total} chips")
        try:
            bet = int(input("Please input how much you would like to bet: "))
            if bet <= chips.total:
                chips.bet = bet
                print(f"You bet {bet}\n")
                break
            else:
                print("Your bet is too large, please try a smaller bet\n")
        except:
            print("You did not input an integer\n")
            
def hit(deck,hand):
    hand.add_card(deck.deal())
#     print(str(hand.cards[-1]))
#     return hand.value

def hit_or_stand(deck,hand):
    global playing
    choice = input("Please choose hit(h) or stand(s): ")
    if choice =='h':
        print("You chose to hit")
        hit(deck,hand)
        print("Your hand is now: " + str(hand))
        print(f"It is worth: {hand.value}\n")
        
    elif choice == 's':
        playing = False
    else:
        print("Please try again")
        
def show_some(player,dealer):
    print("Players hand: " + str(player))
    print("Dealers hand: " + str(dealer.cards[0]) + " Hidden\n")
    
def show_all(player,dealer):
    print("Players hand: " + str(player))
    print("Dealers hand: " + str(dealer) +"\n")
    
def player_busts(player,chips):
    if player.value > 21:
        chips.lose_bet()
    print("The player busts")

def game_over_check(player_hand,dealer_hand,chips):
    '''
    This can be used to check for any win or ending condition in the game and will act accordinly.
    The inputs are player hand, dealer hand, and playyer chip pile
    '''
    global playing
    global player_bust
    if playing:
        if player_hand.value > 21:
            playing = False
            player_bust = True
            chips.lose_bet()
            print("Player busts")
    else:
        if dealer_hand.value > 21:
            chips.win_bet()
            print("Dealer busts")
        elif dealer_hand.value > player_hand.value:
            chips.lose_bet()
            print("Dealer wins")
        elif player_hand.value > dealer_hand.value:
            chips.win_bet()
            print("Player Wins")
        elif player_hand.value == dealer_hand.value:
            chips.bet = 0
            print("Tie")
        
    
```

```python
player_chips = Chips()
while True:
    print("Welcome to Blackjack\n")
    
    the_deck = Deck()
    the_deck.shuffle()
    
    take_bet(player_chips)
    
    player_hand = Hand()
    dealer_hand = Hand()
    
    for i in range(2):
        player_hand.add_card(the_deck.deal())
        dealer_hand.add_card(the_deck.deal())
        
    show_some(player_hand,dealer_hand)
    
    player_bust = False
    playing = True
    while playing:
        hit_or_stand(the_deck,player_hand)
        game_over_check(player_hand,dealer_hand,player_chips)
    
    while player_bust == False:
    
        show_all(player_hand,dealer_hand)

        while dealer_hand.value < 17:
            hit(the_deck,dealer_hand)
            print(dealer_hand)
            print(dealer_hand.value)
            print("")
        
        game_over_check(player_hand,dealer_hand,player_chips)
        
        break
    
    while True:
        answer = input("Would you like to play again (y/n): ")
        if answer == 'y' or answer == 'n':
            break
    if answer == 'n':
        break
    clear_output()
```
